# Retina

````python
item = 'Retina'
df_para_item = df_para.query("mask == '{}'".format(item))
print('\n[run start] {}\n'.format(item))

aligned=True
is_1v1 = False

n_epochs = [100, 200, 200]

for _i, _row in df_para_item.iterrows():
    run_cross_species_models(
        path_adata1=_row['path_ref'],
        path_adata2=_row['path_que'],
        key_class1=_row['key_cell_type'],
        key_class2=_row['key_cell_type'],
        sp1=_row['sp_simple_ref'],
        sp2=_row['sp_simple_que'],
        tissue_name=_row['tissue'],
        aligned=aligned,
        resdir_tag='{}-map-{};epochs={};is_1v1={}'.format(
            _row['name_ref'],_row['name_que'],
            sum(n_epochs),is_1v1),
        resdir=p_res,
        models='csMAHN,came'.split(','),
        **{'n_epochs': n_epochs,
           'is_1v1': is_1v1})

# SAMap
for _i, _row in df_para_item.iterrows():
    run_cross_species_models(
        path_adata1=_row['path_ref'],
        path_adata2=_row['path_que'],
        key_class1=_row['key_cell_type'],
        key_class2=_row['key_cell_type'],
        sp1=_row['sp_simple_ref'],
        sp2=_row['sp_simple_que'],
        tissue_name=_row['tissue'],
        aligned=aligned,
        resdir_tag='{}-map-{};is_1v1={}'.format(
            _row['name_ref'],
            _row['name_que'],
            is_1v1),
        resdir=p_res,
        models='SAMap'.split(','),
        **{
           'is_1v1': is_1v1})
````

````python
item = 'Retina-parameters'
df_para_item = df_para.query("mask == '{}'".format(item))
print('\n[run start] {}\n'.format(item))
df_para_item
aligned = True
is_1v1 = False

n_epochs = [100, 200, 200]

for n_hvgs, n_degs in [
    [1000, 50], [2000, 50], [3000, 50],
        [1000, 100], [2000, 100], [3000, 100]]:
    print('{} {}'.format(n_hvgs, n_degs))
    for _i, _row in df_para_item.iterrows():
        run_cross_species_models(
            path_adata1=_row['path_ref'],
            path_adata2=_row['path_que'],
            key_class1=_row['key_cell_type'],
            key_class2=_row['key_cell_type'],
            sp1=_row['sp_simple_ref'],
            sp2=_row['sp_simple_que'],
            tissue_name=_row['tissue'],
            aligned=aligned,
            resdir_tag='{}-map-{};epochs={};is_1v1={};n_hvgs={};n_degs={}'.format(
                _row['name_ref'], _row['name_que'],
                sum(n_epochs), is_1v1,
                n_hvgs, n_degs),
            resdir=p_res,
            models='csMAHN'.split(','),
            **{'n_epochs': n_epochs,
               'is_1v1': is_1v1,
               'n_hvgs': n_hvgs,
               'n_degs': n_degs
               })
````

````python

for nfeats, hidden in [
    [32, 64],
    [64, 64],  # default
    [128, 64],
    [64, 32],
    [64, 128]

]:
    print('{} {}'.format(nfeats, hidden))
    for _i, _row in df_para_item.iterrows():
        run_cross_species_models(
            path_adata1=_row['path_ref'],
            path_adata2=_row['path_que'],
            key_class1=_row['key_cell_type'],
            key_class2=_row['key_cell_type'],
            sp1=_row['sp_simple_ref'],
            sp2=_row['sp_simple_que'],
            tissue_name=_row['tissue'],
            aligned=aligned,
            resdir_tag='{}-map-{};epochs={};is_1v1={};nfeats={};hidden={}'.format(
                _row['name_ref'], _row['name_que'],
                sum(n_epochs), is_1v1,
                nfeats, hidden),
            resdir=p_res,
            models='csMAHN'.split(','),
            **{'n_epochs': n_epochs,
               'is_1v1': is_1v1,
              'nfeats':nfeats,
               'hidden':hidden})
````
